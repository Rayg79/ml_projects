```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib
import os
```

```python
data.query("country == 'NG'")
```

```python
dmelt = data.melt(['country', 'year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],'sex-age', 'cases')
```

```python
dmelt.head(100)
```

```python
dmelt['sex'] = dmelt['sex-age'].str.slice(0,1)
```

```python
dmelt['age'] = dmelt['sex-age'].str.slice(1)
```

```python
dmelt
```

```python
dmelt['age'] = dmelt['age'].map({'04':'0-4','514':'5-14', '014':'0-14', '1524':'15-24', '2534':'25-34',
               '3544':'35:44', '4554':'45-54', '5564':'55-64', '65': '65+', 'u':np.nan})
```

```python
dmelt
```

```python
dmelt.dropna(subset = ['cases'])
```
