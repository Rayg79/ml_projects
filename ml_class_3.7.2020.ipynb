```python
newData = pd.DataFrame()
for files in adata:
    df = pd.read_csv(files)
    newData = pd.concat([newData, df])
```

```python
newData
```

```python
newData.dropna()
```

```python
bestData = newData[newData['Quantity Ordered'] != 'Quantity Ordered']
```

```python
bestData
```

```python
bestData.dropna(inplace = True)
```

```python
bestData
```

```python
bestData['Quantity Ordered'] = pd.to_numeric(bestData['Quantity Ordered'])
```

```python
bestData['Price Each'] = pd.to_numeric(bestData['Price Each'])
```

```python
bestData['Total Sales'] = bestData['Quantity Ordered'] * bestData['Price Each'] 
```

```python
bestData
```

```python
bestData['Month'] = bestData['Order Date'].str[0:2]
```

```python
bestData
```

```python
bestData['Month'] = pd.to_numeric(bestData['Month'])
```

```python
bestData
```

```python
grpData = bestData.groupby('Month')['Total Sales']
```

```python
MSales = grpData.sum()
month = range(1,13)
```

```python
plt.bar(month, MSales)
plt.xticks(month)
plt.title('Sales Report')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show
```
