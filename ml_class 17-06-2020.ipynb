```python
# the first import is an import of chart plotter from matplotlib 
import matplotlib.pyplot as plt

# converting the numpy import to a short name
import numpy as np

# calling a line for the chart in matplotlib
%matplotlib inline

# calling pillow library that caries image
from PIL import Image

# it is use to read file and copy file path into our project
import glob
```

```python
# declaring a variable pie for the low point of the chart
low = -np.pi # -3.14
low
```

```python
# declaring a variable pie for the high point of the chart
high = np.pi # 3.14
high
```

```python
# declaring the range of itiration from the low to high value
N = 300
```

```python
# declaring range of -3.14 to 3.14 itirating through in 300 places
x = np.linspace(low, high, N)
x
```

```python
# declaring different variable for sin cos and tan simotanously
y = np.sin(x)
z = np.cos(x)
a = np.tan(x)
```

```python
# ploting a chart for (x) range declaration and (y) which is a sin
plt.plot(x,y)
plt.show()
```

```python
# ploting a chart for (x) range declaration and (z) which is a cos
plt.plot(x,z)
plt.show()
```

```python
# ploting a chart for (x) range declaration and (a) which is a tan
plt.plot(x,a)
plt.show()
```
