```python
sales = [file for file in os.listdir('./Sales_Data')]
```

```python
newData = pd.DataFrame()
```

```python
for files in sales:
    df = pd.read_csv('./Sales_Data/' + files)
    newData = pd.concat([newData, df])
```

```python
newData
```

```python
newData.dropna(inplace = True)
```

```python
newData
```

```python
newData['Month'] = newData['Order Date'].str[0:2]
```

```python
newData
```

```python
newData.query("Month == 'Or'")
```

```python
bestData = newData[newData['Month'] != 'Or']
```

```python
bestData
```

```python
bestData['Month'] = pd.to_numeric(bestData['Month'])
```

```python
bestData['Quantity Ordered'] = pd.to_numeric(bestData['Quantity Ordered'])
bestData['Price Each'] = pd.to_numeric(bestData['Price Each'])
```

```python
bestData['Gross Price'] = bestData['Quantity Ordered'] * bestData['Price Each']
```

```python
bestData
```

```python
grpMonth = bestData.groupby('Month')['Gross Price']
```

```python
grpMonth.count()
```

```python
monthSales = grpMonth.sum()
```

```python
month = range(1, 13)
```

```python
plt.bar(month, monthSales)
plt.xticks(month)
plt.grid()
plt.show()
```

```python
def getCity(address):
    return address.split(',')[1]
def getState(address):
    return address.split(',')[2].split(' ')[1]
```

```python
bestData['City'] = bestData['Purchase Address'].apply(lambda x: f'{getCity(x)} ({getState(x)})')
```

```python
bestData
```

```python
grpState = bestData.groupby('City')['Gross Price']
```

```python
grpState.count()
```

```python
unqCity = bestData['City'].unique()
unqCity
```

```python
city = [unqCity for unqCity, bestData in grpState]
```

```python
city
```

```python
plt.plot(city, grpState.count())
plt.xticks(rotation = 90)
plt.show()
```
