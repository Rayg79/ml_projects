```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('who.csv')
```

```python
data
```

```python
ngdata = data.query('country == "NG"')
ngdata
```

```python
ngdata.columns
```

```python
ngmelt = ngdata.melt(['country', 'year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'], 'sex-age', 'cases')
```

```python
ngmelt
```

```python
ngmelt['sex'] = ngmelt['sex-age'].str.slice(0,1)
ngmelt['age'] = ngmelt['sex-age'].str.slice(1)
```

```python
ngmelt
```

```python
ngmelt.drop(['sex-age'], axis = 1, inplace = True)
```

```python
ngmelt.dropna(subset = ['cases'], inplace = True)
```

```python
ngmelt
```

```python
ngcases = ngmelt
```
